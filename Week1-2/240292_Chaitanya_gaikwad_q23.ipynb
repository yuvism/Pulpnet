# Removing stopwords

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
```

```python
### it also tranform text to lowercase as well
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    text = str(text).lower()
    words = text.split()
    clean_words = []

    for word in words:
        if word not in stop_words:
            clean_words.append(word)

    return ' '.join(clean_words)



data['text_without_stopwords'] = data['text'].apply(remove_stopwords)




print(data[['text', 'text_without_stopwords']].sample(4))
```

## Removing Special Characters by python regex

```python
import re
```

```python
def remove_special_characters(text):
    text = str(text)
    cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return cleaned_text

data["text_without_special"]=data["text_without_stopwords"].apply(remove_special_characters)
print(data[['text_without_stopwords', 'text_without_special']].head())
```

in the above code puntuations are also removed ,but if to remove some specific special characters like hashtag etc. we can just insert it in re.sub(pattern,"",text) change the pattern

# Converting all text into lowercase

```python
def convert_to_lower(text):
  text=str(text)
  new_text = text.lower()
  return new_text

data["text_lowercase"]=data["text_without_special"].apply(convert_to_lower)
data.to_csv("updated.csv", index=False)
```

```python

import pandas as pd
import ast
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet


def get_wordnet_pos(treebank_tag: str):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN


def parse_token_column(entry):
    """
    If entry is already a list â†’ return it.
    If entry is a Pythonâ€literal string like "['a','b']" â†’ ast.literal_eval â†’ list.
    If entry is a plain string "a b c" â†’ split on whitespace.
    Otherwise â†’ return [].
    """
    if isinstance(entry, list):
        return entry
    if isinstance(entry, str):
        try:
            parsed = ast.literal_eval(entry)
            if isinstance(parsed, list):
                return parsed
        except (ValueError, SyntaxError):
            pass
        return entry.split()
    return []


def lemmatize_token_list(tokens: list[str]) -> list[str]:
    lemmatizer = WordNetLemmatizer()
    pos_tags = pos_tag(tokens)  
    return [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in pos_tags
    ]


def row_lemmatizer(token_entry) -> list[str]:
    tokens = parse_token_column(token_entry) 
    if not tokens:
        return []
    return lemmatize_token_list(tokens)


sample = ["running", "cats", "better", "dogs"]
print("Sample lemmatization â†’", row_lemmatizer(sample))

```
